```python
from ollama import chat


def get_temperature(city: str) -> str:
    """주어진 도시의 현재 기온을 반환한다"""
    temperatures = {"Seoul": "26°C", "Tokyo": "24°C"}
    return temperatures.get(city, "데이터 없음")


messages = [{"role": "user", "content": "서울(Seoul) 기온이 어때?"}]
response = chat(model='gemma4:e4b', messages=messages, tools=[get_temperature])
print(response)
```

```python
if response.message.tool_calls:
    call = response.message.tool_calls[0]
    print('모델이 요청한 함수:', call.function.name, call.function.arguments)
    result = get_temperature(**call.function.arguments)

    messages.append(response.message)
    messages.append({'role': 'tool', 'tool_name': call.function.name, 'content': str(result)})

    final_response = chat(model='gemma4:e4b', messages=messages, tools=[get_temperature])

    print(final_response)
```
